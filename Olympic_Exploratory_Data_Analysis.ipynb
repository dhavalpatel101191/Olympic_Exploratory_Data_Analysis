```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

/----------------------Load the Dataset----------------------------/

```python
Athlete = pd.read_csv('./athlete_events.csv')
Regions = pd.read_csv('./noc_regions.csv')
```

```python
Athlete.head()
```

```python
Regions.head()
```

/---------------------Join the Dataframe---------------------------/

```python
Athlete_Regions = pd.merge(left=Athlete,right=Regions,how='left',on='NOC')
```

```python
Athlete_Regions.head()
```

/----------------------------Rename Columns-----------------------/

```python
Athlete_Regions.rename(columns={
    'region':'Region',
    'notes':'Notes'
})
```

/----------------------------Check For Null Values-------------------/

```python
Athlete_Regions.info()
```

```python
# Check which columns has Null values
Athlete_Regions.isna().any()
```

```python
#Check how many Null values in columns
Athlete_Regions.isna().sum()
```

```python
# Print the Columns name containing null Vallues or Missing values in the Form of List.
Athlete_List = Athlete_Regions.columns[Athlete_Regions.isnull().any()].tolist()
Athlete_List
```

/----------------Details of Particular Country-------------------------/

```python
Athlete_Regions[Athlete_Regions['region'] == 'India']
```

```python
Athlete_Regions[Athlete_Regions['region'] == 'USA']
```

/-------------------------------Top 10 Countries Participating---------------------------/

```python
Top_10_Contries_Participate = Athlete_Regions['Team'].value_counts().sort_values(ascending=False).head(10)
Top_10_Contries_Participate
```

/------------------------Plot for Top_10_Contries_Participate ------------/

```python
plt.figure(figsize=(12,6))
plt.title('Top 10 Contries by Participant')
sns.barplot(x=Top_10_Contries_Participate.index,y=Top_10_Contries_Participate,palette='Set3')
```

/-------------------------Age Distribution of the Participants----------------/

```python
plt.figure(figsize=(12,6))
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participant')
plt.hist(Athlete_Regions.Age, bins = np.arange(10,80,2),color='pink',edgecolor='black')
```

/----------------Summer and Winter Olympics Sports----------------/

```python
Winter_Sport = Athlete_Regions[Athlete_Regions.Season == 'Winter'].Sport.unique()
Summer_Sport = Athlete_Regions[Athlete_Regions.Season == 'Summer'].Sport.unique()
```

/-----------------Male and Female Participants----------------------/

```python
Gender_Counts = Athlete_Regions.Sex.value_counts()
Gender_Counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution of Athletes')
plt.pie(Gender_Counts, labels=Gender_Counts.index, autopct='%1.1f%%',startangle=180, shadow=True)
```

/---------------------------------Total Medals--------------------/

```python
Athlete_Regions.Medal.value_counts()
```

/------------------------------Total Number of Females Atheletics in Olympics----------------/

```python
Female_Athlete = Athlete_Regions.loc[
    ((Athlete_Regions.Sex == 'F') & (Athlete_Regions.Season == 'Summer')), ['Sex','Year']
    ].groupby('Year').count().reset_index()
```

```python
Female_Athlete
```

```python
Female_Olympic = Athlete_Regions[(Athlete_Regions.Sex == 'F') & (Athlete_Regions.Season == 'Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=Female_Olympic, palette='Spectral')
plt.title('Female Contributation')
```

```python
Total_Female = Female_Olympic.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
Total_Female[:, 'F'].plot()
plt.title('Female Athelete Over time')
```

/----------------------------Gold Medalist Atheletics-------------------------/

```python
Athlete_Regions.loc[Athlete_Regions.Medal == 'Gold']
```

```python
# Gold Medal at Age Greater than 60
Over_60 = Athlete_Regions.loc[((Athlete_Regions.Medal == 'Gold') & (Athlete_Regions.Age > 60)), ['Sport']]
Over_60
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(Over_60['Sport'])
plt.title('Medal Over 60 of Age')
```

/------------------------Gold Medal For Each Country---------------------/

```python
Total_Gold_Medals = Athlete_Regions.loc[Athlete_Regions.Medal == 'Gold'].Team.value_counts().reset_index(name='Total Gold Medals')
Total_Gold_Medals
```

```python
Plot = sns.catplot(x='index',y='Total Gold Medals',data=Total_Gold_Medals.head(10),height=5,kind='bar',palette='rocket')
Plot.despine(left=True)
Plot.set_xlabels('Top 10 Countries')
Plot.set_ylabels('Number of Medals')
plt.title('Gold Medals by Country')
```

/-------------------------------Rio 2016 Olympic Data------------------------------/

```python
Max_Year = Athlete_Regions.Year.max()
Rio_Gold = Athlete_Regions[(Athlete_Regions.Year == Max_Year) & (Athlete_Regions.Medal == 'Gold')].Team.value_counts().head(10)
Rio_Gold
```

```python
sns.barplot(x=Rio_Gold,y=Rio_Gold.index)
plt.ylabel(None)
plt.xlabel('Country With Gold Medals in Rio')
```
